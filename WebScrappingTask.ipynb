```python
url_format = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
import pandas as pd
books_df = pd.DataFrame(columns=['book_title', 'book_price', 'book_rate', 'book_availability'])
books_df
```

```python
book_titles = []
book_prices =[]
book_rates = []
book_availability_lst = []
```

```python
page_no = 1
# Loop through multiple pages
num = 1
for num in range(21):
    url = url_format.format(page_no)

    response = requests.get(url)

    html_content = response.content

    soup = BeautifulSoup(html_content, 'html.parser')

    books = soup.find_all('article', class_='product_pod')
    for i in range(len(books)):
        book_title = books[i].find('h3').text.strip()
        book_titles.append(book_title)
        book_price = books[i].find('div', class_='product_price').find('p',class_='price_color').text.strip()
        book_prices.append(book_price)
        book_rate = books[i].find('p', class_='star-rating')['class'][-1]
        book_rates.append(book_rate)
        book_availability = books[i].find('div', class_='product_price').find('p',class_='instock availability').text.strip()
        book_availability_lst.append(book_availability)
        print(book_title, book_price, book_rate, book_availability)
    
    page_no +=1
```

```python
books_df['book_title'] = book_titles
books_df['book_price'] = book_prices
books_df['book_rate'] = book_rates
books_df['book_availability'] = book_availability_lst
books_df
```

```python
books_df.to_csv('Books.csv')
```
